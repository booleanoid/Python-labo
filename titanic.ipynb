Nullデータの穴埋め

```python
data['Sex'].replace(['male', 'female'], [0, 1], inplace=True)
```

```python
data['Sex']
```

```python
data['Embarked'].fillna(('S'), inplace=True)
data['Embarked'] = data['Embarked'].map({'S':0, 'C':1, 'Q':2})
data['Embarked']
```

```python
fare_mean = np.mean(data['Fare'])
fare_mean
```

```python
data['Fare'].fillna((fare_mean), inplace=True)
data['Fare']
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=data['Age'], kde=False, bins=8)
plt.show()
```

```python
print('平均値：',data['Age'].mean())
print('中央値：',data['Age'].median())
```

```python
age_avg = data['Age'].mean()
age_std = data['Age'].std()

data['Age'].fillna(np.random.randint(age_avg - age_std, age_avg + age_std), inplace=True)
data['Age']
```

```python
delete_columns = ['Name', 'PassengerId', 'SibSp', 'Parch', 'Ticket', 'Cabin']
data.drop(delete_columns, axis=1, inplace=True)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
train = data[~data['Survived'].isnull()]
test  = data[data['Survived'].isnull()]
```

```python
test.head()
```

```python
y_train = train['Survived']
X_train = train.drop('Survived', axis=1)
X_test = test.drop('Survived', axis=1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(penalty='l2', solver='sag', random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_pred[:20]
```
